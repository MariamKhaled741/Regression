```python
def remove_outliers_iqr(df, columns):
    df_clean = df.copy()
    for col in columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    return df_clean
```

```python

numeric_cols = [
    'Rooms', 'Distance', 'Postcode', 'Bedroom2', 'Bathroom', 'Car',
    'Landsize', 'Lattitude', 'Longtitude', 'Propertycount', 'price'
]

clean_df = remove_outliers_iqr(full_df, numeric_cols)
```

```python
import numpy as np
import pandas as pd
import scipy.stats
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.model_selection import train_test_split

class MultipleLinearRegression:
    def __init__(self):
        """
        Initializes the Multiple Linear Regression model.
        """
        self.coefficients = None
        self.MSE = None
        self.r_squared = None
        self.n = None
        self.p = None  # number of features

    def fit(self, X, y):
        """
        Fits the multiple linear regression model to the training data.
        """
        self.X = X
        self.y = y
        X = np.array(X)
        y = np.array(y)
        X = np.c_[np.ones((X.shape[0], 1)), X]  # Add intercept

        self.n, self.p = X.shape
        XTX = X.T @ X
        XTX_inv = np.linalg.inv(XTX)
        XTy = X.T @ y
        self.coefficients = XTX_inv @ XTy

        y_hat = self.predict(X[:, 1:])  # exclude intercept
        y_bar = y.mean()

        self.SSE = sum((y - y_hat) ** 2)
        self.SST = sum((y - y_bar) ** 2)
        self.SSR = sum((y_hat - y_bar) ** 2)

        self.MSE = self.SSE / (self.n - self.p)
        self.r_squared = self.SSR / self.SST

    def predict(self, X):
        """
        Predicts the target variable using the fitted model.
        """
        if self.coefficients is None:
            raise ValueError("The model has not been fitted yet.")

        X = np.array(X)
        X = np.c_[np.ones((X.shape[0], 1)), X]  # Add intercept
        return X @ self.coefficients

    def anova_table(self):
        """
        Constructs and returns the ANOVA table for the model.
        """
        if self.coefficients is None:
            raise ValueError("The model has not been fitted yet.")

        df_regression = self.p - 1
        df_error = self.n - self.p
        df_total = self.n - 1

        MSR = self.SSR / df_regression
        MSE = self.SSE / df_error
        self.F_stat = MSR / MSE

        data = {
            'Source': ['Regression', 'Error', 'Total'],
            'Sum of Squares': [self.SSR, self.SSE, self.SST],
            'Degrees of Freedom': [df_regression, df_error, df_total],
            'Mean Square': [MSR, MSE, ""],
            'F-Statistic': [self.F_stat, "", ""]
        }

        return pd.DataFrame(data)

    
    def hypothesis_test(self, alpha=0.05):
        """
        Perform a hypothesis test to check model significance.
        """
        if self.coefficients is None:
            raise ValueError("The model has not been fitted yet.")

        df_regression = self.p - 1
        df_error = self.n - self.p
        F_c = scipy.stats.f.ppf(1 - alpha, df_regression, df_error)

        if self.F_stat > F_c:
            conclusion = (
                "Since F_0 > F_c, we reject the null hypothesis.\n"
                "Therefore, the model is significant (at least one predictor is useful)."
            )
        else:
            conclusion = (
                "Since F_c > F_0, we don't reject the null hypothesis.\n"
                "Therefore, the model is not significant."
            )

        print("=" * 50)
        print(" Hypothesis Testing Results (Overall Model) ")
        print("=" * 50)
        print(f"{'Null Hypothesis (H_0):':<30} All coefficients (except intercept) = 0")
        print(f"{'Alternative Hypothesis (H_a):':<30} At least one coefficient ≠ 0")
        print("-" * 50)
        print(f"{'F-statistic (F_0):':<30} {self.F_stat:.4f}")
        print(f"{'Critical value (F_c):':<30} {F_c:.4f}")
        print("-" * 50)
        print(conclusion)
        print("=" * 50)
def coefficient_tests(self, alpha=0.05):
        """
        Perform t-tests for individual coefficients.
        """
        if self.coefficients is None:
            raise ValueError("The model has not been fitted yet.")

        X = np.c_[np.ones((self.n, 1)), self.X]
        XTX_inv = np.linalg.inv(X.T @ X)
        se = np.sqrt(np.diag(XTX_inv) * self.MSE)

        t_stats = self.coefficients / se
        p_values = 2 * (1 - scipy.stats.t.cdf(np.abs(t_stats), self.n - self.p))

        t_critical = scipy.stats.t.ppf(1 - alpha/2, self.n - self.p)
        lower_ci = self.coefficients - t_critical * se
        upper_ci = self.coefficients + t_critical * se

        results = {
            'Coefficient': self.coefficients,
            'Std Error': se,
            't-stat': t_stats,
            'p-value': p_values,
            f'{int((1 - alpha) * 100)}% CI Lower': lower_ci,
            f'{int((1 - alpha) * 100)}% CI Upper': upper_ci
        }

        return pd.DataFrame(results, index=['Intercept'] + [f'X{i}' for i in range(1, self.p)])

    def plot_residuals(self):
        """
        Plot residuals vs fitted values.
        """
        if self.coefficients is None:
            raise ValueError("The model has not been fitted yet.")

        y_hat = self.predict(self.X)
        residuals = self.y - y_hat

        plt.figure(figsize=(10, 5))
        plt.scatter(y_hat, residuals, alpha=0.6)
        plt.axhline(y=0, color='r', linestyle='--')
        plt.xlabel('Fitted values')
        plt.ylabel('Residuals')
        plt.title('Residuals vs Fitted Values Plot')
        plt.show()

    def plot_3d(self, X, y):
        """
        Plot the relationship between input features (X) and target variable (y) in a 3D plot.
        """
        if self.coefficients is None:
            raise ValueError("The model has not been fitted yet. Please call the fit() function first.")

        if X.shape[1] > 2:
            X = X[:, :2]  # Use only the first two features if there are more

        fig = plt.figure(figsize=(10, 7))
        ax = fig.add_subplot(111, projection='3d')

        ax.scatter(X[:, 0], X[:, 1], y, color='blue', label='Actual data')

        x1_range = np.linspace(X[:, 0].min(), X[:, 0].max(), 50)
        x2_range = np.linspace(X[:, 1].min(), X[:, 1].max(), 50)
        x1_mesh, x2_mesh = np.meshgrid(x1_range, x2_range)
        X_mesh = np.c_[x1_mesh.ravel(), x2_mesh.ravel()]
        X_mesh_with_intercept = np.c_[np.ones(X_mesh.shape[0]), X_mesh]  # Add intercept
        y_mesh = X_mesh_with_intercept @ self.coefficients  # Predictions from the model

        y_mesh = y_mesh.reshape(x1_mesh.shape)

        ax.plot_surface(x1_mesh, x2_mesh, y_mesh, color='r', alpha=0.3)

        ax.set_xlabel('X1')
        ax.set_ylabel('X2')
        ax.set_zlabel('Price')
        ax.set_title('3D Plot of Multiple Linear Regression')

        plt.legend()
        plt.show()

    def interval_estimation(self, X=None, alpha=0.05):
        """
        Estimate confidence intervals for coefficients and predictions.
        """
        if self.coefficients is None:
            raise ValueError("The model has not been fitted yet. Please call the fit() function first.")

        t_value = scipy.stats.t.ppf(1 - alpha/2, self.n - self.p)

        X_with_intercept = np.c_[np.ones((self.n, 1)), self.X]
        XTX_inv = np.linalg.inv(X_with_intercept.T @ X_with_intercept)
        std_errors = np.sqrt(np.diag(XTX_inv) * self.MSE)

        coef_intervals = [
            [coef - t_value * std, coef + t_value * std]
            for coef, std in zip(self.coefficients, std_errors)
        ]

        pred_intervals = None
        if X is not None:
            X = np.array(X)
            if X.ndim == 1:
                X = X.reshape(1, -1)

            X_with_intercept_pred = np.c_[np.ones((X.shape[0], 1)), X]
            pred_std_errors = np.array([
                np.sqrt(x.T @ XTX_inv @ x) * np.sqrt(self.MSE)
                for x in X_with_intercept_pred
            ])

            predictions = self.predict(X)
            pred_intervals = [
                [pred - t_value * std, pred + t_value * std]
                for pred, std in zip(predictions, pred_std_errors)
            ]

        return coef_intervals, pred_intervals
```
